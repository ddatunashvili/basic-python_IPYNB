```python
'სიასთან მუშაობა'
ten_things = "ვაშლი ფორთოხალი ყვავი ტელეფონი შუქი შაქარი"

print("მოიცადეთ, ამ სიაში 10 ნივთი არაა. მოდით გამოვაწოროთ.")

stuff = ten_things.split(' ') # შექმნის სიას
more_stuff = ["დღე", "ღამე", "სიმღერა", "თამაში", "სიმინდი", "ბანანი", "გოგო", "ბიჭი"]
while len(stuff) != 10:
    next_one = more_stuff.pop()
    print("დამატეთ: ", next_one)
    stuff.append(next_one)
    print(f"ახლა სიაში გვაქვს {len(stuff)} ნივთი.")
    
print("აი ახლა ვნახოთ ჩვენი სია: ", stuff)

print("მოდით კიდევ რამე გავაკეთოთ.")

print(stuff[1])
print(stuff[-1]) 
print(stuff.pop())
print(' '.join(stuff)) # აქცევს სტრინგად
print('#'.join(stuff[3:5]))
a=' '.join(stuff) # ცვლადი შეიცავს სტრინგს
print(a)
```
