```python
#                                 კომპოზიცია
'იგი იძლევა საშუალებას ავირიდოთ მულტი შთამომავლობა'
# სხვა გზაც არსებობს,სხვა კლასების და მოდულების გამოყენება იმპლიცისტურს ჯობია
# მოდულიდან ფუნქციების გამოძახება ჯობია
#-------------------------------------------------
'  მაგალითად:  '
class მამიკო(object):           
    def უარი(self):       
        print('მშობლის უარყოფა')
    def იმპლიცისტური(self):
        print('იმპლიცისტური მშობელი')# პირველი >> დაიბეჭდა მშობლისგან აღებული ფუნქცია
    def შეცვლა(self):
        print('შეცვლილი მშობელი >> მეოთხე') # მეოთხე >> დაიბეჭდა მშობლის ფუნქცია
# თავისი თვისებები არაქვს და იძახებს სხვა კლასის თვისებებს
# არ აქვს შთამომავლობა
class შვილი(object):
    def __init__(self):
        # გააჩნია პარამეტრი მამა (დროებითი წარმომადგენელი)
        self.მამა = მამიკო()
    def უარი(self): 
        print('შვილის უარი')  # მეორე >> დაიბეჭდა შვილის ფუნქცია
    def იმპლიცისტური(self):
        self.მამა.იმპლიცისტური() 
    def შეცვლა(self):
        print('შვილი, სანამ მშობელი შეიცვლება >> მესამე') # მესამე >> დაიბეჭდა შვილის ფუნქცია
        self.მამა.შეცვლა()
        print('შვილი, მშობლის შეცვლის შემდგომ >> მეხუთე') # მეხუთე >> დაიბეჭდა შვილის ფუნქცია
ბიჭი = შვილი()
print('-- პირველი --')
ბიჭი.იმპლიცისტური() 
print('--- მეორე --')
ბიჭი.უარი()
print('--- მესამე --')
ბიჭი.შეცვლა()
print('============')


```

```python
                    'იერარქია ზევიდან ქვევით'
'''
შვილს აქვს a თვისება დედისგან, 
    დედას აქვს a თვისება პაპასგან
შვილს აქვს b თვისება მამისგან, 
    მამას აქვს b თვისება ბებოსგან
'''    
    
class ბებო(object):
    print('მე ვარ ბებო')
    def b(self): # იპოვა b
        print('ბებო: უიმე ქაა მოხვედიი')
class პაპა(object):
    print('მე ვარ პაპა')
    def a(self): # იპოვა a
        print('პაპა: პაპიი მოხვედი პაპაი შვილო') 
class მამა(ბებო,პაპა): # შვილი პირველი მშობელისკენ წავიდა
    print('მე ვარ მამა')
    pass
class დედა(ბებო,პაპა):
    print('მე ვარ დედა')
    pass
class შვილი(დედა,მამა):
    def სახელი(self):
        print('მე ვარ შვილი') 
    def a(self):
        print('ვეძებ a')
        super(შვილი,self).a() # წადი დედასთან a
    def b(self):
        print('ვეძებ b')
        super(შვილი,self).b() # წადი დედასთან b:x ; წადი მამასთან b
ბიჭი = შვილი()
ბიჭი.სახელი()
ბიჭი.a()
ბიჭი.b()
```

```python
'იერარქია ქვევიდან ზევით'
'''
შვილს აქვს a თვისება დედისგან, 
    დედას აქვს a თვისება პაპასგან
შვილს აქვს b თვისება მამისგან, 
    მამას აქვს b თვისება ბებოსგან
'''    
   
class ბებო1(object):
    def a(self): # იპოვა b
        print('ბებო1: უიმე ქაა მოხვედიი')
class პაპა1(object):
    def b(self): # იპოვა a
        print('პაპა1: პაპიი მოხვედი პაპაი შვილო')
class ბებო2(object):
    def a(self): # იპოვა b
        print('ბებო2: უიმე ქაა მოხვედიი')
class პაპა2(object):
    def b(self): # იპოვა a
        print('პაპა2: პაპიი მოხვედი პაპაი შვილო') 
class მამა(პაპა1,ბებო1):
    pass
class დედა(პაპა2,ბებო2):
    pass
class შვილი(დედა,მამა):
    def s(self):
        print('მე ვარ შვილი')
    def a(self):
        super(მამა, self).a()
        super(მამა, self).b()
        super(დედა, self).b()
        super(დედა, self).a()
ბიჭი = შვილი()
ბიჭი.s()
ბიჭი.a()

```
