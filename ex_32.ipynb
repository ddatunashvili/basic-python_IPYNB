```python
'სიები და ციკლები'

count=[1,2,3,4,5]
f=['ა','ბ','გ','დ']
c=[1,'ა', 2, 'ბ',3, 'გ',4,'დ']
# for ციკლი
for i in count:
    print(i,'რიცხვები')
for e in f:
    print(e,'ასოები')
for o in c:
    print(o,'შერეული')
# ცარიელი სიის შექმნა
e=[]
# ჩაციკლე 0 დან 5 მდე
for i in range(0,6):
    print(i,'0 დან 5 მდე')
    # ჩაამარე e სიაში ყოველი ელემენტი
    e.append(i)
for i in e:
    print(i,'e ში ელემენტები')
```

```python
# შეიქმნა ორ ელემენტიანი სია
a=[0] + [1]
print(a)
a=[0] + ['1']
print(a)
# გვაქვს დამატება და გამრავლება
a=[0] + [4]*2 # ჩაწერე 0 და ორი 4 იანი
print(a)
```

```python
წინადადება='წინადადება'
# ასოებად დაკუწვა
for ასო in წინადადება:
    print(ასო)
print('==============')
# იტერირების შემდეგ არსებობას განაგრძობს
for ასოები in წინადადება:
    print(ასოები, end='')
print(ასოები)
print('==============')
# append ს შეუძლია მხოლოდ ერთი ელემენტის დამატება
#  ობიექტის ჩამატება
el=[1,2,3,4,5]
el.append([6,7])
print(el)
print('==============')
# სიას სიას მიამატებს
el.extend([6,7])
print(el)
print('==============')
# შეტრიალება მიმდევრობის
el.reverse()
print(el)
print('==============')
# ანბანის მიხედვით დაალაგებს
el=['ა','გ','ბ','დ']
el.sort()
print(el)
print('==============')
```

```python
'ფუნქციების ცდა'
სია=[1,3,4,5,6,7,7]
# დათვლის რამდენი 7 იანია
სია.count(7)
# იპოვის 4 იანის ინდექსს
სია.index(4)
# აბრუნებს სიის კოპიას
სია.copy()
# დაამატებს 1 ინდექსზე 2 იანს
სია.insert(1,2)
print(სია)
# შემოატრიალებს 
სია.reverse()
print(სია)
# დააგრძელებს სიასს 8 9 მდე
სია.extend([8,9])
print(სია)
# დაალაგებს სწორად
სია.sort()
print(სია)
# ამოიღებს 7 იანს
სია.remove(7)
print(სია)
# წაშალა მე-8 ინდექსზე და გამოიტანა 9 იანი
სია.pop(8)
# სულ წაშალა სია
სია.clear()
print(სია)
```
