```python
# ჩემი აწყობილი ამოცანა " სამკუთხედის ფართობი "
import math
import time
'''
მოცემულია სამკუთხედი
რომლის გერდების პარამეტრებია:
a = 13
b = 13
c = 10
'''
a = int(input("შეიყვანე a გვერდის სიგრძე")) 
b = int(input("შეიყვანე b გვერდის სიგრძე")) 
c = int(input("შეიყვანე c გვერდის სიგრძე"))
#--------------------------------------
# 
# როგორი სამკუთხედია
if a == b == c:
    print('ტოლგვერდა სამკუთხედია')
    print('ვითვლი ფართობს ')
    for i in range (3): 
        print('მოიცა ..')
        time.sleep(1)
    print("ვიცი, მაგრამ არ გეტყვი :P ")
elif a==b or a == c or b == c:
    print('ტოლფერდა სამკუთხედია')
    # ვიპოვნოთ კათეტები
    ad = cd = int(c / 2)
    print('კათეტები = ', ad,'და', cd)
    # ვიპოვნოთ სიმაღლე
    h = math.sqrt(a**2 - cd **2)
    print("სიმაღლე = ", int(h))
    # სამკუთხედის ფართობი
    S = 1/2 * h * c
    print("სამკუთხედის ფართობი =", int(S))  
else:
    print('სხვადასხვაგვერდა სამკუთხედია')
    print('ვითვლი ფართობს ')
    for i in range (3): 
        print('მოიცა ..')
        time.sleep(1)
    print("ვიცი, მაგრამ არ გეტყვი :P ")
```
