```python
'დავალება'
'''
1. შევცვალე
2. 1 ფუნტი = 0.4535923 კილოგრამი 
   1 ინჩი = 2,54 სანტიმეტრი
3. round() არ ცვლის და იმ მომენტისთვის ამრგვალებს,
დანარჩენი ფუნქციები ახსნილი მაქვსს მაგალითებში
'''
# 2 
# ფუნქციას გადაყავს კილოგრამები ფუნტებში
კილოგრამი_ფუნტი = lambda x: x * 1/0.4535923
x = float(input('შეიყვანე კილოგრამი : ') )
print(კილოგრამი_ფუნტი(x), 'ფუნტი')
# ცვლადს გადაყავს კილოგრამები ფუნტებში
კილოგრამი = float(input('შეიყვანე კილოგრამი :'))
ფუნტი = კილოგრამი * 1/0.4535923
print(ფუნტი, 'ფუნტი')
# ცვლადს გადაყავს სანტიმეტრი ინჩებში
სანტიმეტრი = float(input('შეიყვანე სანტიმეტრი :'))
ინჩი = სანტიმეტრი * 1/ 2.54
print(ინჩი, 'ინჩი')
```
